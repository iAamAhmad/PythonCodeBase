```python
import seaborn as sns
import matplotlib.pyplot as plt

titanic = sns.load_dataset('titanic')
```

```python
print(titanic.head())
```

## $\color{Green}{\text{The majority of passengers embarked At!}}$

```python
# Create a countplot of embarkation port
sns.countplot(x='embark_town', data=titanic)

# Show the plot
plt.show()
```

```python
titanic.to_csv('data')
```

### The plot shows that the majority of passengers embarked at Southampton, with a smaller number embarking at Cherbourg and Queenstown.

## $\color{Green}{\text{Distribution of Passengers by Class}}$
### Let's start by creating a bar chart to show the distribution of passengers by class. This can be useful in understanding the socio-economic background of the passengers and how it may have influenced their likelihood of survival.

```python
sns.countplot(x='class', data=titanic)
plt.title('Distribution of Passengers by Class')
plt.show()
```

###  We can see that the majority of passengers were in third class, followed by first and second class.

## $\color{Green}{\text{Survival Rate by Class and Sex}}$
### Next, let's create a heatmap to show the survival rate of passengers by class and sex. This can help us understand whether certain groups of passengers were more likely to survive than others.

```python
pivot = titanic.pivot_table(index='class', columns='sex', values='survived', aggfunc='mean')
sns.heatmap(pivot, annot=True, cmap='Blues')
plt.title('Survival Rate by Class and Sex')
plt.show()
```

### We can see that the highest survival rate was among female passengers in first and second class. Male passengers in third class had the lowest survival rate.

## $\color{green}{\text{Age Distribution of Passengers by Sex}}$

### Let's create a histogram to show the age distribution of passengers by sex. This can help us understand the demographics of the passengers.

```python
g = sns.FacetGrid(data=titanic, col='sex')
g.map(sns.histplot, 'age', bins=20)
plt.subplots_adjust(top=0.8)
g.fig.suptitle('Age Distribution of Passengers by Sex')
plt.show()
```

### We can see that the age distribution is slightly different between male and female passengers, with a higher concentration of male passengers in their 20s and 30s.

## $\color{green}{\text{Fare Distribution by Class and Survival}}$
### Let's create a box plot to show the fare distribution of passengers by class and survival. This can help us understand whether there is a correlation between fare paid and likelihood of survival.

```python
sns.boxplot(x='class', y='fare', hue='survived', data=titanic)
plt.title('Fare Distribution by Class and Survival')
plt.show()
```

### We can see that in each class, the median fare paid by survivors is higher than that of non-survivors. Additionally, the fare distribution in first and second class is generally higher than that in third class.

## The survival rate of children
### The survival rate of children under 10 years old was relatively high compared to other age groups. This could be due to the "women and children first" protocol that was followed during the evacuation of the ship.

```python
## Load the Titanic dataset
titanic = sns.load_dataset('titanic')

## Add a column for whether the passenger is a child (under 10 years old)
titanic['is_child'] = titanic['age'] < 10

## Compute the survival rate for each age group
age_groups = titanic.groupby(['is_child', 'age']).mean()['survived']

## Create a bar plot of the survival rate by age group
sns.barplot(x='is_child', y='survived', data=titanic)

## Show the plot
plt.show()
```

### This supports the idea that the "women and children first" protocol was followed during the evacuation of the Titanic, as children (and women) were given priority over men in accessing lifeboats.
